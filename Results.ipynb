```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import json
import os
```

```python
json_no_standard_norm_path = r"C:\Users\alber\Desktop\Make Believe Diciembre\Models\Info Dictionaries"

json_standard_norm_path = r"C:\Users\alber\Desktop\Make Believe Diciembre\V2\Dicts"

json_standard_norm_training = r"C:\Users\alber\Desktop\Make Believe Diciembre\V2\Dicts training normalized together, test normalized separately"

list_no_standard_norm = os.listdir(json_no_standard_norm_path)

list_standard_norm_path = os.listdir(json_standard_norm_path)

list_standard_norm_training = os.listdir(json_standard_norm_training)
```

```python
list_dicts_standard_norm_training = []

for file in list_standard_norm_training:
    
    path = str(json_standard_norm_training)+ "\{}".format(file)
    
    # print(path)
    
    with open(path) as f:
        data = json.load(f)
        list_dicts_standard_norm_training.append(data)
```

```python
list_dicts_standard_norm_train_test = []

for file in list_standard_norm_path:
    
    path = str(json_standard_norm_path)+ "\{}".format(file)
    
    # print(path)
    
    with open(path) as f:
        data = json.load(f)
        list_dicts_standard_norm_train_test.append(data)
```

```python
list_dicts_no_standard_norm_no_train_no_test = []

for file in list_no_standard_norm:
    
    path = str(json_no_standard_norm_path)+ "\{}".format(file)
    
    # print(path)
    
    with open(path) as f:
        data = json.load(f)
        list_dicts_no_standard_norm_no_train_no_test.append(data)
        
```

```python
# for item in list_dicts_no_standard_norm_no_train_no_test[0:1]:
#     print(item.keys())
    
for item in list_dicts_no_standard_norm_no_train_no_test:
    if "cv_scores" not in item.keys():
        item["cv_scores"] = "not applicable"
    if "mean_cv_accuracy" not in item.keys():
        item["mean_cv_accuracy"] = "not applicable"
    
```

```python
# for item in list_dicts_standard_norm_train_test[0:1]:
#     print(item.keys())
    
for item in list_dicts_standard_norm_train_test:
    if "cv_scores" not in item.keys():
        item["cv_scores"] = "not applicable"
    if "mean_cv_accuracy" not in item.keys():
        item["mean_cv_accuracy"] = "not applicable"
```

```python
# for item in list_dicts_standard_norm_training[0:1]:
#     print(item.keys())
    
for item in list_dicts_standard_norm_training:
    item["cv_scores"] = "not applicable"
    item["mean_cv_accuracy"] = "not applicable"
```
